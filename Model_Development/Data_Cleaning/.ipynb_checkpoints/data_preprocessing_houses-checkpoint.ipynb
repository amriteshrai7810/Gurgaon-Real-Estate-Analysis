```python
import numpy as np
import pandas as pd
```

```python
# View all rows and columns
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
```

```python
# Import the data
df = pd.read_csv('../Data_Scraping/Data/Gurgaon/Independent House/ind_house.csv')
```

```python
df.sample(5)
```

```python
df.info()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Columns to drop: Link, 
df.drop(columns=['link', 'property_id'], inplace=True)
```

```python
df.head()
```

```python
# Renaming the columns
df.rename(columns={'rate': 'price_per_sqft'}, inplace=True)
df.head()
```

```python
# Society
df['society'].value_counts()
```

```python
df['society'].isnull().sum()
```

```python
df['society'] = df['society'].str.replace('nan', 'independent')
```

```python
df['society'].value_counts().shape  # Checking number of unique socities
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts()
```

```python
# Price
df['price'].value_counts()
```

```python
df['price'].isnull().sum()
```

```python
df[df['price'] == 'Price on Request'].shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
# Create a function to convert all prices in same unit

def normalize_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(normalize_price)
```

```python
df.head()
```

```python
# price per sqft
df['price_per_sqft'].value_counts()
```

```python
# get only price per sqft
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',','').str.strip().astype('float')
```

```python
df.sample(5)
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df.head()
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
df.sample(5)
```

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df[df['bathroom'].isnull()]
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df.sample(5)
```

```python
# balcony
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No', '0')
```

```python
# additional Room
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].value_counts().shape
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'].fillna('not available', inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.sample(5)
```

```python
# floor num
df.rename(columns={'noOfFloor': 'floorNum'}, inplace=True)
```

```python
df.sample(5)
```

```python
df['floorNum'].value_counts().shape
```

```python
df['floorNum'].isnull().sum()
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).str.replace('Ground', '0').str.replace('Basement', '-1').str.replace('Lower', '0').str.extract(r'(\d+)')
```

```python
df.sample(5)
```

```python
# facing
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA', inplace=True)
```

```python
df['facing'].value_counts()
```

```python
df['area'] = round((df['price']*10000000/df['price_per_sqft']))
```

```python
df.insert(loc=1, column='property_type', value='house')
```
